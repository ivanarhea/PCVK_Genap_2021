Jawab :
Int16 Ini adalah FCL type. Dalam C#, pendek dipetakan ke Int16. Int32Ini adalah FCL type.Dalam C#, int dipetakan ke Int32.Ini adalah jenis nilai dan mewakili Sistem.Int32 struct.Hal ini ditandatangani dan membutuhkan 32 bit Memiliki minimal -2147483648 dan maksimum +2147483647 nilai.
